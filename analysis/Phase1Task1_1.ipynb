```python
# Label PR type
pr_df["is_agentic"] = True
human_pr_df["is_agentic"] = False

# Keep only relevant columns
agent_cols = ["id", "title", "body", "merged_at", "is_agentic"]
human_cols = ["id", "title", "body", "merged_at", "is_agentic"]

combined_pr_df = pd.concat(
    [
        pr_df[agent_cols],
        human_pr_df[human_cols]
    ],
    ignore_index=True
)

print("Total PRs:", len(combined_pr_df))
```

```python
combined_pr_df["title"] = combined_pr_df["title"].fillna("")
combined_pr_df["body"] = combined_pr_df["body"].fillna("")

combined_pr_df["text"] = (
    combined_pr_df["title"] + " " + combined_pr_df["body"]
).str.lower()

```

```python
quality_keywords = {

    "Security": ["security", "secure"],
    "User Assistance": ["user assistance", "help system", "documentation", "guide"],
    "Performance Efficiency": ["performance efficiency", "performance", "optimization", "optimize"],
    "Compatibility": ["compatibility", "compatible"],
    "Maintainability": ["maintainability", "maintainable"],
    "Safety": ["safety", "safe"],
    "Risk Identification": ["risk identification", "risk analysis", "threat detection"],
    "Reliability": ["reliability", "reliable"],
    "Interaction Capability": ["interaction capability", "interaction", "user interaction"],
    "Operational Constraint": ["operational constraint", "constraint", "limitation"],
    "Integrity": ["integrity", "data integrity"],
    "Functional Correctness": ["functional correctness", "correctness"],
    "Flexibility": ["flexibility", "flexible"],
    "User Error Protection": ["user error protection", "error prevention"],
    "Co-existence": ["co-existence", "coexistence"],
    "Availability": ["availability", "uptime"],
    "Functional Completeness": ["functional completeness", "complete functionality"],
    "Reusability": ["reusability", "reusable"],
    "Modularity": ["modularity", "modular"],
    "Appropriateness Recognizability": ["appropriateness recognizability"],
    "Testability": ["testability", "test coverage"],
    "Fault Tolerance": ["fault tolerance", "fault tolerant"],
    "Capacity": ["capacity"],
    "Scalability": ["scalability", "scalable"],
    "User Engagement": ["user engagement", "engagement"],
    "Resource Utilization": ["resource utilization", "cpu usage", "memory usage"],
    "Resistance": ["resistance", "resilient"],
    "Adaptability": ["adaptability", "adaptable"],
    "Interoperability": ["interoperability", "integration"],
    "Authenticity": ["authenticity", "authentication"],
    "Self-descriptiveness": ["self-descriptiveness", "self descriptive"],
    "Inclusivity": ["inclusivity", "inclusive"],
    "Hazard Warning": ["hazard warning", "warning system"]
}


import re

quality_patterns = {}

for quality, keywords in quality_keywords.items():
    pattern = r"\b(" + "|".join(re.escape(k) for k in keywords) + r")\b"
    quality_patterns[quality] = re.compile(pattern, re.IGNORECASE)
```

```python
def detect_quality(text):
    detected = []
    for quality, keywords in quality_keywords.items():
        if any(keyword in text for keyword in keywords):
            detected.append(quality)
    return detected

combined_pr_df["quality_attributes"] = (
    combined_pr_df["text"].apply(detect_quality)
)
```

```python
def detect_quality(text):
    detected = []

    if not isinstance(text, str):
        return detected

    for quality, pattern in quality_patterns.items():
        if pattern.search(text):
            detected.append(quality)

    return list(set(detected))
```

```python
combined_pr_df["quality_attributes"] = (
    combined_pr_df["text"].apply(detect_quality)
)
```

```python
combined_pr_df["merged"] = ~combined_pr_df["merged_at"].isna()

combined_pr_df["has_quality"] = (
    combined_pr_df["quality_attributes"].apply(lambda x: len(x) > 0)
)
```

```python
acceptance_rate = (
    combined_pr_df.groupby("has_quality")["merged"].mean()
)

acceptance_rate
```

```python
from scipy.stats import chi2_contingency

contingency = pd.crosstab(
    combined_pr_df["has_quality"],
    combined_pr_df["merged"]
)

chi2, p, dof, expected = chi2_contingency(contingency)

print("Chi-square:", chi2)
print("p-value:", p)
```
