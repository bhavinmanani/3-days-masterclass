```python
# Label PR type
pr_df["is_agentic"] = True
human_pr_df["is_agentic"] = False

# Keep only relevant columns
agent_cols = ["id", "title", "body", "merged_at", "is_agentic"]
human_cols = ["id", "title", "body", "merged_at", "is_agentic"]

combined_pr_df = pd.concat(
    [
        pr_df[agent_cols],
        human_pr_df[human_cols]
    ],
    ignore_index=True
)

print("Total PRs:", len(combined_pr_df))
```

```python
combined_pr_df["title"] = combined_pr_df["title"].fillna("")
combined_pr_df["body"] = combined_pr_df["body"].fillna("")

combined_pr_df["text"] = (
    combined_pr_df["title"] + " " + combined_pr_df["body"]
).str.lower()

```

```python
quality_keywords = {
    "Security": ["security", "authentication", "authorization",
                 "confidentiality", "integrity", "encryption",
                 "privacy", "vulnerability"],

    "Performance efficiency": ["performance", "optimize", "latency",
                               "speed", "memory", "cpu", "throughput"],

    "Usability": ["usability", "ux", "ui", "accessibility"],

    "Reliability": ["bug", "failure", "crash", "error", "fault"],

    "Maintainability": ["refactor", "clean code", "modular",
                        "technical debt"],

    "Portability": ["cross-platform", "portable", "windows", "linux"],

    "Compatibility": ["compatibility", "backward compatible"],

    "Functional suitability": ["feature", "functionality", "requirement"]
}
```

```python
def detect_quality(text):
    detected = []
    for quality, keywords in quality_keywords.items():
        if any(keyword in text for keyword in keywords):
            detected.append(quality)
    return detected

combined_pr_df["quality_attributes"] = (
    combined_pr_df["text"].apply(detect_quality)
)
```

RQ1: How frequently are ISO/IEC 25010 quality characteristics
mentioned in agentic pull requests compared to the human pull
request?

```python
trace_links = (
    combined_pr_df
    .explode("quality_attributes")
    [["id", "quality_attributes"]]
    .dropna()
)

trace_links.columns = ["PR_id", "quality_attribute"]

trace_links.to_csv("trace_links.csv", index=False)

trace_links.head()
```

```python
rq1_df = combined_pr_df.explode("quality_attributes")

freq = (
    rq1_df
    .groupby(["is_agentic", "quality_attributes"])
    .size()
    .unstack()
    .fillna(0)
)

freq
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Sort columns for consistency
freq = freq.sort_index(axis=1)

qualities = freq.columns
x = np.arange(len(qualities))

width = 0.35

fig, ax = plt.subplots(figsize=(12,6))

agentic_counts = freq.loc[True] if True in freq.index else np.zeros(len(qualities))
human_counts = freq.loc[False] if False in freq.index else np.zeros(len(qualities))

ax.bar(x - width/2, agentic_counts, width, label="Agentic PRs")
ax.bar(x + width/2, human_counts, width, label="Human PRs")

ax.set_ylabel("Number of Mentions")
ax.set_title("RQ1: ISO 25010 Mentions in Agentic vs Human PRs")
ax.set_xticks(x)
ax.set_xticklabels(qualities, rotation=45, ha="right")
ax.legend()

plt.tight_layout()
plt.show()
```

```python
freq_percent = freq.div(freq.sum(axis=1), axis=0) * 100

freq_percent.T.plot(kind="bar", figsize=(12,6))
plt.ylabel("Percentage (%)")
plt.title("RQ1: Percentage of Quality Mentions (Agentic vs Human)")
plt.xticks(rotation=45, ha="right")
plt.legend(["Human PRs", "Agentic PRs"])
plt.tight_layout()
plt.show()
```

RQ2: Which quality characteristics frequently co-occur?

```python
from itertools import combinations
from collections import Counter

co_occurrence = Counter()

for qualities in combined_pr_df["quality_attributes"]:
    if len(qualities) > 1:
        for pair in combinations(sorted(qualities), 2):
            co_occurrence[pair] += 1

co_df = (
    pd.DataFrame(co_occurrence.items(), columns=["Pair", "Count"])
    .sort_values("Count", ascending=False)
)

co_df.head(10)
```

```python
import seaborn as sns

heatmap_df = pd.DataFrame(0,
                          index=quality_keywords.keys(),
                          columns=quality_keywords.keys())

for (q1, q2), count in co_occurrence.items():
    heatmap_df.loc[q1, q2] = count
    heatmap_df.loc[q2, q1] = count

plt.figure(figsize=(8,6))
sns.heatmap(heatmap_df, annot=True, fmt="d")
plt.title("Quality Attribute Co-occurrence")
plt.show()
```

RQ4 â€” Correlation with Acceptance

```python
combined_pr_df["merged"] = ~combined_pr_df["merged_at"].isna()

combined_pr_df["has_quality"] = (
    combined_pr_df["quality_attributes"].apply(lambda x: len(x) > 0)
)
```

```python
acceptance_rate = (
    combined_pr_df.groupby("has_quality")["merged"].mean()
)

acceptance_rate
```

```python
from scipy.stats import chi2_contingency

contingency = pd.crosstab(
    combined_pr_df["has_quality"],
    combined_pr_df["merged"]
)

chi2, p, dof, expected = chi2_contingency(contingency)

print("Chi-square:", chi2)
print("p-value:", p)
```
